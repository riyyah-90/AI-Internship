```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split, GridSearchCV, cross_val_score
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.impute import SimpleImputer
```

```python
df = pd.read_csv('Dataset.csv')
```

```python
df['sex'] = df['sex'].astype(str).replace({"Male": 0, "Female": 1})
df['target'] = df['target'].astype(str).replace({"False": 0, "True": 1})
df
```

```python
# Check for missing values
print("\nMissing Values:")
missing_values = df.isnull().sum()
print(missing_values)
```

```python
#Getting rid of rows with NaN values
df = df.dropna()
df
```

```python
# Separate features and target
x = df.drop('target', axis=1)
y = df['target']
```

```python
x_encoded = pd.get_dummies(x)
x_encoded
```

```python
x_train, x_test, y_train, y_test = train_test_split(x_encoded, y, test_size=0.2, random_state=42)
```

```python
# Scaling the Data
scaler = StandardScaler()
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)
```

```python
# Convverting to DataFrame
x_train_scaled = pd.DataFrame(x_train_scaled, columns=x_train.columns)
x_test_scaled = pd.DataFrame(x_test_scaled, columns=x_test.columns)
```

```python
x_train_scaled.shape
```

```python
# K-Nearest Neighbour
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# Initialize and train the KNN model
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(x_train_scaled, y_train)

# Make predictions
y_pred = knn.predict(x_test_scaled)

# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
report = classification_report(y_test, y_pred)
matrix = confusion_matrix(y_test, y_pred)

print(f"Accuracy: {accuracy}")
```

```python
# Logistic Regression
from sklearn.linear_model import LogisticRegression

# Initialize and train the Logistic Regression model
lg = LogisticRegression(random_state=42)
lg.fit(x_train_scaled, y_train)

# Make predictions
y_pred_lg = lg.predict(x_test_scaled)

# Evaluate the model
accuracy_lg = accuracy_score(y_test, y_pred_lg)
report_lg = classification_report(y_test, y_pred_lg)
matrix_lg = confusion_matrix(y_test, y_pred_lg)

print(f"Logistic Regression Accuracy: {accuracy_lg}")
```

```python
# Decision tree classifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# Initialize and train the Decision Tree model
dtc = DecisionTreeClassifier(random_state=42)
dtc.fit(x_train_scaled, y_train)

# Make predictions
y_pred_dtc = dtc.predict(x_test_scaled)

# Evaluate the model
accuracy_dtc = accuracy_score(y_test, y_pred_dtc)
report_dtc = classification_report(y_test, y_pred_dtc)
matrix_dtc = confusion_matrix(y_test, y_pred_dtc)

print(f"Decision Tree Classifier Accuracy: {accuracy_dtc}")
```

```python
# Linear support vector classifier
from sklearn.svm import LinearSVC
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# Initialize and train the Linear SVC model
lsvc = LinearSVC(random_state=42)
lsvc.fit(x_train_scaled, y_train)

# Make predictions
y_pred_lsvc = lsvc.predict(x_test_scaled)

# Evaluate the model
accuracy_lsvc = accuracy_score(y_test, y_pred_lsvc)
report_lsvc = classification_report(y_test, y_pred_lsvc)
matrix_lsvc = confusion_matrix(y_test, y_pred_lsvc)

print(f"Linear SVC Accuracy: {accuracy_lsvc}")
```

```python
# Kernelized Support Vector Classifier
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# Initialize and train the SVC model with a kernel (e.g., 'rbf')
svc = SVC(kernel='rbf', random_state=42) # You can try other kernels like 'linear', 'poly', 'sigmoid'
svc.fit(x_train_scaled, y_train)

# Make predictions
y_pred_svc = svc.predict(x_test_scaled)

# Evaluate the model
accuracy_svc = accuracy_score(y_test, y_pred_svc)
report_svc = classification_report(y_test, y_pred_svc)
matrix_svc = confusion_matrix(y_test, y_pred_svc)

print(f"Kernelized SVC Accuracy: {accuracy_svc}")
```

```python
# Random Forest Classifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# Initialize and train the Random Forest model
rfc = RandomForestClassifier(random_state=42)
rfc.fit(x_train_scaled, y_train)

# Make predictions
y_pred_rfc = rfc.predict(x_test_scaled)

# Evaluate the model
accuracy_rfc = accuracy_score(y_test, y_pred_rfc)
report_rfc = classification_report(y_test, y_pred_rfc)
matrix_rfc = confusion_matrix(y_test, y_pred_rfc)

print(f"Random Forest Classifier Accuracy: {accuracy_rfc}")
```
